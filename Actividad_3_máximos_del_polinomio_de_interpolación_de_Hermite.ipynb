```
import numpy as np
import scipy.optimize as optimize
import scipy.interpolate as interpolate
import pandas as pd
import matplotlib.pyplot as plt
```

```
#Cargar los datos

df=pd.DataFrame({'X':[2.5,3.1,4.5,5,5.9,6.2],
              'Y':[3,-1.01,1.5,0.7,2.8,1.5]})
```

```
#Interpolación de Hermit

H=np.polynomial.hermite.Hermite.fit(df.X,df.Y,5)

#Graficar
x=np.linspace(df.X.iloc[0],df.X.iloc[-1],100)
plt.plot(x,H(x))
plt.plot(df.X,df.Y,'ro')
plt.xlabel('X')
plt.ylabel('Y')
```

```
#Para hallar los máximos de H(x) podemos calcular los mínimos de -H(x)

plt.plot(x,H(x),'b-',label = "$H(x)$")
plt.plot(x,-H(x),'c-',label = "$-H(x)$")
plt.plot(df.X,df.Y,'ro')
plt.legend()
plt.grid()
plt.xlabel('X')
plt.ylabel('Y')
```

```
#Minimos de -H(x)

min1=optimize.fmin_powell(-H,4.5,full_output=True)
min2=optimize.fmin_powell(-H,5.7,full_output=True)
```

```
#Se encuentra que la función optimize.fmin_powell no está identificando correctamente el mínimo local.
#Lo hace correctamente con el mínimo global de -H(x)
#Se hicieron pruebas para valores desde 3.5 hasta 4.5 y no se obtuvo resultado para el mínimo local.
```
