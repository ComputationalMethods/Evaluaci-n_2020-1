Resolver las ecuaciones diferenciales$$
\frac{d}{d t} \alpha_{a}^{-1}=-\frac{b_{a}}{2 \pi} \quad(a=1,2,3),
$$donde, para un valor inicial del parametro, $t$, $t_0=4.5$ (adimensional)$$
\left(\alpha_{1}, \alpha_{2}, \alpha_{3}\right)=(0.01694, 0.03375, 0.1176),
$$en dos casos:$$
\left(b_{1}, b_{2}, b_{3}\right)=\left\{\begin{array}{ll}
(41 / 10,-19 / 6,-7) &amp; \text { Standard Model } \\
(33 / 5,1,-3) &amp; \text { MSSM }
\end{array}\right.
$$Demostrar que en el rango: $t_0$ a $t_{\text{fin}}=50$ solo en el segundo caso: MSSM, las tres funciones $\alpha_a^-1$ convergen a un punto, y determinar el correspondiente valor de $t$.

```
import numpy as np
import scipy.integrate as integrate
import matplotlib.pyplot as plt
```

```
#Definición variables

(α1,α2,α3)=(0.01694,0.03375,0.1176)
(α1_inv,α2_inv,α3_inv)=(1/α1,1/α2,1/α3)
α_inv0 =[α1_inv,α2_inv,α3_inv]

#Standard model

(b1_s,b2_s,b3_s)=(41/10,-19/6,-7)
bs = np.array([b1_s,b2_s,b3_s])

#MSSM

(b1_m,b2_m,b3_m)= (33/5,1,-3)
bm = np.array([b1_m,b2_m,b3_m])

#Definición del tiempo
t0 = 4.5
tfin = 50
t = np.linspace(t0,tfin,100)
```

```
#Caso 1 Standard model

def dainv_dt_s(a_inv,t):
    z = a_inv
    dz = -bs/(2*np.pi)
    return dz

#Sol ec diferencial
a_s = integrate.odeint(dainv_dt_s,α_inv0,t)
```

```
#Caso 2 MSSM

def dainv_dt_m(a_inv,t):
    z = a_inv
    dz = -bm/(2*np.pi)
    return dz

#Sol ec diferencial
a_m = integrate.odeint(dainv_dt_m,α_inv0,t)
```

```
#Gráficamente

plt.figure(figsize=(17.5,6))
plt.subplot(1,2,1)
plt.title('Standard Model')
plt.plot(t,a_s[:,0],'c-',label="$α_{1}^{-1}$")
plt.plot(t,a_s[:,1],'k:',label="$α_{2}^{-1}$")
plt.plot(t,a_s[:,2],'b--',label="$α_{3}^{-1}$")
plt.ylabel('$α^{-1}$',size=15)
plt.ylim(10,60)
plt.grid()
plt.legend()

plt.subplot(1,2,2)
plt.title('MSSM')
plt.plot(t,a_m[:,0],'c-',label="$α_{1}^{-1}$")
plt.plot(t,a_m[:,1],'k:',label="$α_{2}^{-1}$")
plt.plot(t,a_m[:,2],'b--',label="$α_{3}^{-1}$")
plt.vlines(37.5,8,70,colors='grey',linestyles='dashdot')
plt.xlabel('t',size=15)
plt.ylabel('$α^{-1}$',size=15)
plt.ylim(10,60)
plt.grid()
plt.legend()

plt.show()


```

Las líneas aparentan cruzarse en un punto cercano a t = 37.5 y ésto sólo ocurre en el modelo MSSM.
