```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.DataFrame()

m = 0.5
g = 9.8
Fd = [0.00, 0.01, 0.05]

for c in Fd:

    x = np.array([0, 50, 0])
    v = np.array([5, 0, 0])
    p = m*v
    t = 0
    deltat = 0.01
    
    df = df.append({'x':x, 'p':p, 't':t, 'c':c}, ignore_index=True)
    
    for t in np.arange(0, 3, deltat):
        F = np.array([0, -m*g, 0]) - c*p**2
        p = p + F*deltat
        x = x + (p/m)*deltat
        t = t + deltat
        df = df.append({'x':x, 'p':p, 't':t, 'c':c}, ignore_index=True)
```

```
plt.plot(df[df.c==Fd[0]].x.str[0], df[df.c==Fd[0]].x.str[1], label='$Fd={}$'.format(Fd[0]))
plt.plot(df[df.c==Fd[1]].x.str[0], df[df.c==Fd[1]].x.str[1], label='$Fd={}$'.format(Fd[1]))
plt.plot(df[df.c==Fd[2]].x.str[0], df[df.c==Fd[2]].x.str[1], label='$Fd={}$'.format(Fd[2]))
plt.xlabel('$x$', size=20)
plt.ylabel('$y$', size=20)
plt.grid()
plt.legend()
plt.show()
```
