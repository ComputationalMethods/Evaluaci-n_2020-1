Resolver las ecuaciones diferenciales
$$
\frac{d}{d t} \alpha_{a}^{-1}=-\frac{b_{a}}{2 \pi} \quad(a=1,2,3),
$$
donde, para un valor inicial del parametro, $t$, $t_0=4.5$ (adimensional) 
$$
\left(\alpha_{1}, \alpha_{2}, \alpha_{3}\right)=(0.01694, 0.03375, 0.1176),
$$
en dos casos:
$$
\left(b_{1}, b_{2}, b_{3}\right)=\left\{\begin{array}{ll}
(41 / 10,-19 / 6,-7) & \text { Standard Model } \\
(33 / 5,1,-3) & \text { MSSM }
\end{array}\right.
$$
Demostrar que en el rango: $t_0$ a $t_{\text{fin}}=50$ solo en el segundo caso: MSSM, las tres funciones $\alpha_a^-1$ convergen a un punto, y determinar el correspondiente valor de $t$. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
# Definición de parámetros y valores iniciales.

α = np.array([0.01694, 0.03375, 0.1176])

b_sm = np.array([41/10, -19/6, -7])*(-1) / (2*np.pi)
b_mssm = np.array([33/5, 1, -3])*(-1) / (2*np.pi)

y0 = α**-1
t = np.linspace(4.5, 50)
```

```python
# Definición de funciones para cada modelo.

def f_sm(y, t):
    return b_sm

def f_mssm(y, t):
    return b_mssm
```

```python
# Valores de α para cada t.

α_sm = integrate.odeint(f_sm, y0, t)
α_mssm = integrate.odeint(f_mssm, y0, t)
```

```python
plt.figure(figsize=(20, 6))

# Gráfica del SM.
plt.subplot(121)
plt.plot(t, α_sm[:, 0], label='α1')
plt.plot(t, α_sm[:, 1], label='α2')
plt.plot(t, α_sm[:, 2], label='α3')
plt.title('SM', size=22)
plt.xlabel('$t$', size=18)
plt.ylabel('$α^{-1}$', size=18)
plt.legend()
plt.grid()

# Gráfica del MSSM.
plt.subplot(122)
plt.plot(t, mssm[:, 0], label='α1')
plt.plot(t, mssm[:, 1], label='α2')
plt.plot(t, mssm[:, 2], label='α3')
plt.title('MSSM', size=22)
plt.xlabel('$t$', size=18)
plt.ylabel('$α^{-1}$', size=18)
plt.legend()
plt.grid()

plt.show()
```

```python
# Las funciones convergen, aproximadamente, en t = 37.6.

plt.plot(t, mssm[:, 0], label='α1')
plt.plot(t, mssm[:, 1], label='α2')
plt.plot(t, mssm[:, 2], label='α3')
plt.title('MSSM', size=22)
plt.xlabel('$t$', size=18)
plt.ylabel('$α^{-1}$', size=18)
plt.legend()
plt.grid()
plt.xlim(36, 40)
plt.ylim(22, 28)

plt.show()
```
